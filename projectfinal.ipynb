```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("roadconstruction.csv")
df1 = df.drop_duplicates()
df1
```

```python
num_cols = [
    'NO_OF_ROAD_WORK_SANCTIONED', 'NO_OF_BRIDGES_SANCTIONED',
    'NO_OF_ROAD_WORKS_COMPLETED', 'NO_OF_BRIDGES_COMPLETED',
    'NO_OF_ROAD_WORKS_BALANCE', 'NO_OF_BRIDGES_BALANCE',
    'LENGTH_OF_ROAD_WORK_SANCTIONED_KM', 'COST_OF_WORKS_SANCTIONED_LAKHS',
    'LENGTH_OF_ROAD_WORK_COMPLETED_KM', 'EXPENDITURE_OCCURED_LAKHS',
    'LENGTH_OF_ROAD_WORK_BALANCE_KM'
]
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')  
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
df['STATE_NAME'] = df['STATE_NAME'].fillna(df['STATE_NAME'].mode()[0])
df['DISTRICT_NAME'] = df['DISTRICT_NAME'].fillna(df['DISTRICT_NAME'].mode()[0])
df['PMGSY_SCHEME'] = df['PMGSY_SCHEME'].fillna(df['PMGSY_SCHEME'].mode()[0])
print("\nAfter Preprocessing:")
print(df.info())
print("\nMissing values after handling:\n", df.isnull().sum())
df.to_csv("pmgsy_cleaned.csv", index=False)
```

```python
#3.To compare the sanctioned and completed number of road and bridge works across PMGSY schemes.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("roadconstruction.csv")


summary = df.groupby('PMGSY_SCHEME')[[
    'NO_OF_ROAD_WORK_SANCTIONED',
    'NO_OF_ROAD_WORKS_COMPLETED',
    'NO_OF_BRIDGES_SANCTIONED',
    'NO_OF_BRIDGES_COMPLETED'
]].sum().reset_index()

road_melted = summary.melt(
    id_vars='PMGSY_SCHEME',
    value_vars=['NO_OF_ROAD_WORK_SANCTIONED', 'NO_OF_ROAD_WORKS_COMPLETED'],
    var_name='STATUS',
    value_name='COUNT'
)
road_melted['WORK_TYPE'] = 'ROAD'

bridge_melted = summary.melt(
    id_vars='PMGSY_SCHEME',
    value_vars=['NO_OF_BRIDGES_SANCTIONED', 'NO_OF_BRIDGES_COMPLETED'],
    var_name='STATUS',
    value_name='COUNT'
)
bridge_melted['WORK_TYPE'] = 'BRIDGE'


combined = pd.concat([road_melted, bridge_melted])


combined['STATUS'] = combined['STATUS'].str.replace('NO_OF_', '').str.replace('_', ' ').str.title()

plt.figure(figsize=(14, 7))
sns.barplot(
    data=combined,
    x='PMGSY_SCHEME',
    y='COUNT',
    hue='STATUS',
    palette='plasma'
)
plt.title("Sanctioned vs Completed Road and Bridge Works across PMGSY Schemes")
plt.xlabel("PMGSY Scheme")
plt.ylabel("Number of Works")
plt.legend(title="Status")
plt.tight_layout()
plt.show()

```

```python

#4.To assess financial performance by comparing sanctioned costs vs. actual expenditures and utilization percentage by state

financials = df.groupby('STATE_NAME')[['COST_OF_WORKS_SANCTIONED_LAKHS', 'EXPENDITURE_OCCURED_LAKHS']].sum().reset_index()

financials['UTILIZATION_PERCENT'] = (financials['EXPENDITURE_OCCURED_LAKHS'] / financials['COST_OF_WORKS_SANCTIONED_LAKHS']) * 100


financials = financials.sort_values(by='UTILIZATION_PERCENT', ascending=False)

plt.figure(figsize=(14, 8))
bar_width = 0.4
states = financials['STATE_NAME']

plt.barh(states, financials['COST_OF_WORKS_SANCTIONED_LAKHS'], height=bar_width, label='Sanctioned Cost', color='skyblue')
plt.barh(states, financials['EXPENDITURE_OCCURED_LAKHS'], height=bar_width * 0.8, label='Actual Expenditure', color='seagreen')

plt.xlabel("Amount (in Lakhs)")
plt.title("Financial Performance: Sanctioned vs Actual Expenditure by State")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#utilization percentage by state
plt.figure(figsize=(10, 6))
sns.heatmap(financials.set_index("STATE_NAME")[['UTILIZATION_PERCENT']], annot=True, cmap="YlGnBu", fmt=".1f")
plt.title("Utilization Percentage by State")
plt.xlabel("Utilization %")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```

```python
#5.To Identify Districts with Outlier Sanctioned Project Costs Using IQR Analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("roadconstruction.csv")  
Q1 = df['COST_OF_WORKS_SANCTIONED_LAKHS'].quantile(0.25)
Q3 = df['COST_OF_WORKS_SANCTIONED_LAKHS'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[
    (df['COST_OF_WORKS_SANCTIONED_LAKHS'] < lower_bound) |
    (df['COST_OF_WORKS_SANCTIONED_LAKHS'] > upper_bound)
]


print("Top Outlier Districts (by cost):")
print(outliers[['STATE_NAME', 'DISTRICT_NAME', 'COST_OF_WORKS_SANCTIONED_LAKHS']]
      .sort_values(by='COST_OF_WORKS_SANCTIONED_LAKHS', ascending=False).head(10))

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['COST_OF_WORKS_SANCTIONED_LAKHS'], color='purple')
plt.title("Outliers in Cost of Works Sanctioned (Lakhs)")
plt.xlabel("Cost of Works Sanctioned (Lakhs)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#6.Pie Chart - Completed vs Balance Road Work
plt.figure(figsize=(6, 6))
completed = df['LENGTH_OF_ROAD_WORK_COMPLETED_KM'].sum()
balance = df['LENGTH_OF_ROAD_WORK_BALANCE_KM'].sum()
plt.pie(
    [completed, balance],
    labels=['Completed', 'Balance'],
    autopct='%1.1f%%',
    colors=['#ff6289', '#ff93ac'],
    startangle=140
)
plt.title("Overall Road Work Status")
plt.tight_layout()
plt.show()
```

```python
#7.To identify regions with the highest number of pending projects 

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("roadconstruction.csv")
df['PENDING_ROAD_WORKS'] = df['NO_OF_ROAD_WORK_SANCTIONED'] - df['NO_OF_ROAD_WORKS_COMPLETED']

df['PENDING_BRIDGE_WORKS'] = df['NO_OF_BRIDGES_SANCTIONED'] - df['NO_OF_BRIDGES_COMPLETED']


pending_summary = df.groupby('PMGSY_SCHEME')[['PENDING_ROAD_WORKS', 'PENDING_BRIDGE_WORKS']].sum().reset_index()

pending_summary['TOTAL_PENDING_PROJECTS'] = pending_summary['PENDING_ROAD_WORKS'] + pending_summary['PENDING_BRIDGE_WORKS']


pending_summary_sorted = pending_summary.sort_values('TOTAL_PENDING_PROJECTS', ascending=False)

print(pending_summary_sorted[['PMGSY_SCHEME', 'TOTAL_PENDING_PROJECTS']].head(10))

plt.figure(figsize=(18, 8))
sns.pointplot(
    data=pending_summary_sorted.head(10),
    x='PMGSY_SCHEME',
    y='TOTAL_PENDING_PROJECTS',
    color='#A73662'
)
plt.title("Pending Projects by PMGSY Scheme")
plt.xlabel("PMGSY Scheme")
plt.ylabel("Total Pending Projects")
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#8.The correlation matrix for the Road and Bridge Work counts
correlation_matrix = df[['NO_OF_ROAD_WORK_SANCTIONED', 
                         'NO_OF_ROAD_WORKS_COMPLETED', 
                         'NO_OF_BRIDGES_SANCTIONED', 
                         'NO_OF_BRIDGES_COMPLETED']].corr()


print(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix for Road and Bridge Work Counts")
plt.tight_layout()
plt.show()
```
